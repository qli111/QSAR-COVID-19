```python
df8=pd.read_csv('bioactivity.csv')
```

```python
df8
```

```python
selection=['bioactivity']
df8[selection]
```

```python
df8_selection = df8[selection]
df8_selection
```

Combine structure and bioactivity

```python
df10 = pd.concat([df4_selection, df8_selection], axis=1)
df10
```

Classification of bioactivities

```python
bioactivity_class = []
for i in df10.bioactivity:
  if float(i) >= 7:
    bioactivity_class.append("active")
  elif float(i) <= 6:
    bioactivity_class.append("inactive")
  else:
    bioactivity_class.append("intermediate")
```

Combine 

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df11 = pd.concat([df10, bioactivity_class], axis=1)
df11
```

Save file as csv file, here we kept index

```python
df12 = df11[df11.bioactivity_class != 'intermediate']
df12
```

```python
df12.to_csv('sars-cov-2-local_bioactivity_data_preprocessed.csv')
```
