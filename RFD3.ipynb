```python
!python <<'PY'
import json, copy

TARGET_PDB = "/content/drive/MyDrive/RFD3/targets/target_CXCR.pdb"
out = "/content/drive/MyDrive/RFD3/inputs/my_CXCR_3ECL.json"

CHAIN = "A"
CONTIG = "30-60,/0,A1-342"

# ä½ è¦çš„ä¸‰ä¸ª ECL åŒºé—´
RANGES = {
    "ECL_1_33": (1, 33),
    "ECL_80_101": (80, 101),
    "ECL_169_191": (169, 191),
}

# æ¯ç§æ°¨åŸºé…¸ä¼˜å…ˆé€‰å“ªäº›â€œä¾§é“¾ä»£è¡¨åŽŸå­â€ï¼ˆå°½é‡é€‰ç¨³å®š/è¡¨å¾æ€§çš„é‡åŽŸå­ï¼‰
# ä½ ä¹Ÿå¯ä»¥æŒ‰éœ€æ±‚æ”¹ï¼Œæ¯”å¦‚æ›´åæžæ€§/å¸¦ç”µåŽŸå­
PREF = {
    "ALA": ["CB"],
    "VAL": ["CG1","CG2"],
    "LEU": ["CD1","CD2"],
    "ILE": ["CD1","CG1","CG2"],
    "SER": ["OG"],
    "THR": ["OG1","CG2"],
    "CYS": ["SG"],
    "MET": ["SD","CE"],
    "PRO": ["CG","CD"],
    "PHE": ["CZ","CE1","CE2"],
    "TYR": ["OH","CZ","CE1","CE2"],
    "TRP": ["CZ2","CH2","CZ3","NE1"],
    "ASN": ["OD1","ND2"],
    "GLN": ["OE1","NE2"],
    "ASP": ["OD1","OD2"],
    "GLU": ["OE1","OE2"],
    "HIS": ["NE2","ND1","CE1"],
    "LYS": ["NZ"],
    "ARG": ["CZ","NH1","NH2","NE"],
    "GLY": [],  # Gly æ²¡ä¾§é“¾ï¼ŒåŽé¢ä¼šå›žé€€åˆ° CA
}

def read_residue_atoms(pdb_path, chain_id):
    """è¿”å›žï¼š{resseq:int -> (resname:str, atoms:set[str])}"""
    out = {}
    with open(pdb_path, "r") as f:
        for line in f:
            if not (line.startswith("ATOM") or line.startswith("HETATM")):
                continue
            # PDB å›ºå®šåˆ—è§£æžæ›´ç¨³
            atom_name = line[12:16].strip()
            resname   = line[17:20].strip()
            ch        = line[21].strip()
            if ch != chain_id:
                continue
            # è·³è¿‡æ°´ç­‰
            if resname in ("HOH", "WAT"):
                continue
            try:
                resseq = int(line[22:26].strip())
            except:
                continue
            # è®°å½•
            if resseq not in out:
                out[resseq] = [resname, set()]
            out[resseq][1].add(atom_name)
    return out

res_atoms = read_residue_atoms(TARGET_PDB, CHAIN)

def atoms_for_res(resname, available_atoms):
    pref = PREF.get(resname, [])
    chosen = [a for a in pref if a in available_atoms]
    # å¦‚æžœä¾§é“¾åŽŸå­ç¼ºå¤±ï¼ˆæˆ–Glyï¼‰ï¼Œå°±å›žé€€åˆ° CA
    if not chosen:
        if "CA" in available_atoms:
            return "CA"
        # æžç«¯æƒ…å†µä¸‹ CA ä¹Ÿæ²¡æœ‰ï¼Œå°±ç”¨ä»»ä½•ä¸€ä¸ªåŽŸå­é¡¶ä¸Šï¼ˆå°½é‡ä¸ä¸ºç©ºï¼‰
        return ",".join(sorted(list(available_atoms))[:1]) if available_atoms else "CA"
    # ä¸ºäº†æ›´å¼ºçº¦æŸï¼Œé€šå¸¸ 1-2 ä¸ªåŽŸå­è¶³å¤Ÿï¼›è¿™é‡Œæœ€å¤šå–2ä¸ª
    chosen = chosen[:2]
    return ",".join(chosen)

BASE = {
    "dialect": 2,
    "infer_ori_strategy": "hotspots",
    "input": TARGET_PDB,
    "contig": CONTIG,
    "is_non_loopy": True,
}

outd = {}
for name, (s,e) in RANGES.items():
    x = copy.deepcopy(BASE)
    hs = {}
    missing = []
    for i in range(s, e+1):
        if i not in res_atoms:
            missing.append(i)
            continue
        resname, atoms = res_atoms[i]
        hs[f"{CHAIN}{i}"] = atoms_for_res(resname, atoms)
    x["select_hotspots"] = hs
    outd[name] = x
    if missing:
        print(f"[WARN] {name}: residues missing in PDB chain {CHAIN}: {missing[:10]}{'...' if len(missing)>10 else ''}")

json.dump(outd, open(out, "w"), indent=2)
print("wrote:", out)
print("tasks:", list(outd.keys()))
print("example select_hotspots (first 10):", list(outd["ECL_1_33"]["select_hotspots"].items())[:10])
PY
```
